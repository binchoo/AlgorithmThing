```python
import random

def swap(arr, a, b) :
    tmp = arr[a]
    arr[a] = arr[b]
    arr[b] = tmp

def quick_sort(arr, p, q) :
    if p < q :
        #get pivot
        k = random.randint(p, q)
        pivot = arr[k]

        swap(arr, p, k)

        left_idx = [i for i in range(p+1, q+1) if arr[i] < pivot]
        right_idx = [i for i in range(p+1, q+1) if arr[i] >= pivot]
        # to do correct answer
        if len(left_idx) < len(right_idx) :
            swaps = len(left_idx)
        else :
            swaps = len(right_idx)

        for i in range(swaps) :
            if left_idx[i] > right_idx[i] :
                swap(arr, left_idx[i], right_idx[i])

        smaller_idx = p
        for i in range(p+1, q+1) :
            if arr[i] < pivot :
                smaller_idx = i
        swap(arr, p, smaller_idx)
        print(arr, pivot)
        quick_sort(arr, p, smaller_idx - 1)
        quick_sort(arr, smaller_idx + 1, q)
```

```python
import numpy as np
arr = np.random.randint(1, 100, 50)
```

```python
quick_sort(arr, 0, len(arr)- 1)
```
