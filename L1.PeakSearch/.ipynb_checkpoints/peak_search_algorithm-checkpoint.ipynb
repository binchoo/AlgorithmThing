```c
#include <stdio.h>


int is_global_maximum(int idx, int* arr, int length)
{
    int a, b, c;
    b = arr[idx];
    if (idx == 0) {
        c = arr[idx + 1];
        return b >= c;
    } else if (idx == length - 1) {
        a = arr[idx - 1];
        return b >= a;
    } else {
        a = arr[idx - 1];
        c = arr[idx + 1];
        return b >= a && b >= c;
    }
    return 0;
}

void
linear_mark_global_maximum(int* arr, int* mark, int length)
{
    for(int i = 0; i < length; i++) {
        mark[i] = is_global_maximum(i, arr, length);
    }
}

#define DTYPE int
#define RTNTYPE DTYPE

RTNTYPE local_max_binary_search(DTYPE* arr, unsigned long length) {
    
    unsigned long middle = (int) (length / 2);
    unsigned long left_length, right_length;
    
    DTYPE* left_arr = arr;
    DTYPE* right_arr = arr + middle + 1;
    
    left_length = middle;
    right_length = length - middle - 1;
    
    if (!(left_length > 0 && right_length > 0)) {
        RTNTYPE base_calc = arr[middle];
        return base_calc;
    }
    if (left_length > 0 && arr[middle - 1] > arr[middle]) {
        RTNTYPE left_calc = local_max_binary_search(left_arr, left_length);
        return left_calc;
    }
    if (right_length > 0 && arr[middle + 1] > arr[middle]) {
        RTNTYPE right_calc = local_max_binary_search(right_arr, right_length);
        return right_calc;
    }   
}

RTNTYPE platform_binary_access(DTYPE* arr, unsigned long length) {
    
    unsigned long middle = (int) (length / 2);
    unsigned long left_length, right_length;
    
    DTYPE* left_arr = arr;
    DTYPE* right_arr = arr + middle + 1;
    
    left_length = middle;
    right_length = length - middle - 1;
    
    if (!(left_length > 0 && right_length > 0)) {
        RTNTYPE base_calc;
        return base_calc;
    }
    if (left_length > 0) {
        RTNTYPE left_calc = platform_binary_access(left_arr, left_length);
        return left_calc;
    }
    if (right_length > 0) {
        RTNTYPE right_calc =platform_binary_access(right_arr, right_length);
        return right_calc;
    }   
}

int main()
{

    int arr[] = {1, 2, 3, 7, 3, 2, 1, 0};
    for(int i = 0; i < sizeof(arr)/sizeof(int); i++) {
        printf("arr[%d] : %d \n", i, is_global_maximum(i, arr, sizeof(arr)/sizeof(int)));
    }
    
    int find = local_max_binary_search(arr, sizeof(arr)/sizeof(int));
    
    printf("local_maximum found : {%d}\n", find);
    return 0;
}
```
